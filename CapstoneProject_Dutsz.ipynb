```python
df_main = df_dataset[['playCount', 'diggCount', 'commentCount', 'shareCount', 'videoMeta/duration', 'musicMeta/musicName', 'createTime']]
print(df_main.head())
```

```python
kolom_yang_dihapus = [
    'effectStickers/0/ID', 'effectStickers/0/name', 'effectStickers/0/stickerStats/useCount',
    'effectStickers/1/ID', 'effectStickers/1/name', 'effectStickers/1/stickerStats/useCount',
    'mentions/0', 'mentions/1', 'mentions/2', 'mentions/3', 'mentions/4', 'mentions/5',
    'hashtags/0/cover', 'hashtags/1/cover', 'hashtags/0/title', 'hashtags/1/title',
    'authorMeta/signature', 'musicMeta/musicAlbum', 'webVideoUrl', 'mediaUrls/0'
]

df_bersih = df_dataset.drop(columns=kolom_yang_dihapus)

df_bersih.info()
df_bersih.head()
```

```python
plt.figure(figsize=(12, 7))

sns.scatterplot(data=df_bersih, x='videoMeta/duration', y='diggCount')
plt.title('Hubungan antara durasi dengan jumlah like')
plt.xlabel('Durasi vi deo (detik)', fontsize=12)
plt.ylabel('Jumlah like', fontsize=12)
plt.show
```

```python
plt.figure(figsize=(12, 7))

sns.scatterplot(data=df_bersih, x='videoMeta/duration', y='commentCount')
plt.title('Hubungan antara durasi dengan jumlah komentar')
plt.xlabel('Durasi video (detik)', fontsize=12)
plt.ylabel('Jumlah komentar', fontsize=12)
plt.show
```

```python
top10_tiktokers = df_bersih['authorMeta/name'].value_counts().nlargest(10).index

df_top10 = df_bersih[df_bersih['authorMeta/name'].isin(top10_tiktokers)]

plt.figure(figsize=(12 ,8))

sns.barplot(data=df_top10, x='authorMeta/name', y='diggCount', palette='viridis')
plt.title('Top 10 TikToker dengan Jumlah Like Tertinggi')
plt.xlabel('Nama TikToker', fontsize=12)
plt.ylabel('Jumlah like', fontsize=12)
plt.show
```

```python
top_10_music = df_bersih['musicMeta/musicName'].value_counts().nlargest(10).index

df_top_10_music = df_bersih[df_bersih['musicMeta/musicName'].isin(top_10_music)]

plt.figure(figsize=(12, 8))

sns.barplot(data=df_top_10_music, x='musicMeta/musicName', y='diggCount', palette='plasma')
plt.title('Top 10 Musik dengan Jumlah Like Tertinggi')
plt.xlabel('Nama Musik', fontsize=12)
plt.ylabel('Jumlah like', fontsize=12)
plt.tight_layout()
plt.show
```

```python
df_bersih['engagement_rate'] = (df_bersih['diggCount'] / (df_bersih['playCount'] + 1)) * 100

top_10_engagement = df_bersih.sort_values(by='engagement_rate', ascending=False).head(10)

print("--- Top 10 Video dengan Engagement Rate (Likes per View) Tertinggi ---")
print(top_10_engagement[['authorMeta/name', 'musicMeta/musicName', 'playCount', 'diggCount', 'engagement_rate']])
```

```python
from langchain_community.llms import Replicate
import os
from google.colab import userdata
# Set the API token
api_token = userdata.get('api_token')
os.environ["REPLICATE_API_TOKEN"] = api_token
# Model setup
model = "ibm-granite/granite-3.3-8b-instruct"
output = Replicate(
model=model,
replicate_api_token=api_token,
)
```

```python
parameters = {
"top_k": 1,
"top_p": 1.0,
"max_tokens": 256,
"min_tokens": 0,
"random_seed": None,
"repetition_penalty": 1.0,
"stopping_criteria": "length (256 tokens)",
"stopping_sequence": None
}
```

```python
insights_text = """ Temuan utama dari analisis data video viral TikTok:
1. Durasi video adalah faktor fundamental. Konten di bawah 60 detik secara konsisten mendominasi engagement, baik dari sisi likes maupun komentar.
2. Popularitas kreator dapat menjadi pengganda kekuatan. Kreator seperti Khaby Lame berada di level viralitas yang jauh lebih tinggi dibandingkan kreator populer lainnya, menunjukkan adanya efek "Winner-Takes-All".
3. Metrik "Engagement Rate" (Likes per View) menunjukkan kualitas viralitas yang sesungguhnya. Ditemukan bahwa video dari kreator yang lebih kecil seringkali memiliki engagement rate lebih tinggi, menandakan koneksi yang lebih kuat dengan audiensnya.
"""
prompt_template = f"""
Summarize the following findings from a TikTok data analysis project into three concise bullet points for a presentation slide. Make the language professional and impactful.

Findings:
{insights_text}

Summary:
"""
```

```python
response = output.invoke(insights_text, parameters)
print(response)
```
